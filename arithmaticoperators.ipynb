```python
#Arithmatic Operators
# +, -, *, /, %
#By default the value of the input value is string in python if you want to change it,s type you have to convert it first


num1=int(input("Enter Number 1: "))
num2=int(input("Enter Number 2: "))

print("Result of addition is : ",num1+num2)

print("Result of subtraction is: ",num1-num2)

print("Result of Multiplcation  is: ",num1*num2)

print("Result of Division is: ",num1/num2)
#If we want to remove the point valu
Num=num1+num2

print("Result of Division is: ",num1//num2)

print("Result of Moulous is: ",num1%num2)




```

```python
#Comparison Operators

print(3>3)

print(5<3)

print(5<=6)

print(5>=2)

print(5!=4)

#Logical Operators
# AND Operator

# NOT Operator
# Only True when both the Conditions are True

print(5>3 and 3<2)
print(2<3 and 3>2)

# OR Operator
#Will be true if one of the conditions will be true

print(5>3 or 3<2)

print(5<6 or 3<2)







```
