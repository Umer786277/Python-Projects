```python
#There are three ways to declare the string first one wthi single quots,double quots,and triple cquots

#Declaring Strngs With Single Quotes
string='NAEEMJANI'
print("Result of the string is: ",string)

#Declaring Strings with Double Quotes
string2="NaeemJni"
print("Result of the string is:",string2)

#Declaring of the Strings with triple Quotes
string3='''Kashif Shehzad'''
print("Result of the third string is: ",string3)
```

```python
#Indexing of the Strings
#First of all i want to make it clear that indexing of the strings in python starts from zero instead of 1 and ends at (N=Total number of items - 1) 
print("Wll shows the result of Index 0: : ",string[0])
print("Will shows the Results :",string[0:4])
#Shows all the items in the Strings
print("Will shows the results: ",string[:])
print("Results are: ", string[0:])
print("Results are: ", string[3:])
```

Negative Indexing 
Negative -1 shows last item in the string
Negative -2 shows second last item in the string and so on

```python
print("Result is :",string2[0:-1])
print("Result is : ",string2[2:-2])
```

```python
#Reversing a python string

print("Reverse string is: ", string3[::-1])
```

```python

# Python Program for
# Formatting of Strings
 
# Default order
String1 = "{} {} {}".format('Geeks', 'For', 'Life')
print("Print String in default order: ")
print(String1)
 
# Positional Formatting
String1 = "{1} {0} {2}".format('Geeks', 'For', 'Life')
print("\nPrint String in Positional order: ")
print(String1)
 
# Keyword Formatting
String1 = "{l} {f} {g}".format(g='Geeks', f='For', l='Life')
print("\nPrint String in order of Keywords: ")
print(String1)
```

```python

# String alignment
String1 = "|{:<10}|{:^10}|{:>10}|".format('Geeks',
                                          'for',
                                          'Geeks')
print("\nLeft, center and right alignment with Formatting: ")
print(String1)
 
# To demonstrate aligning of spaces
String1 = "\n{0:^16} was founded in {1:<4}!".format("GeeksforGeeks",
                                                    2009)
print(String1)
```
